```python
#Converts the text recognized in each notice to string using OCR's tesseract, then splits that text on " "
# to find the buttons representing consent and decline. Not the learn more button as that is recognized in the
# scraper as those learn more buttons have to lead you to another link. 
#problem for accept button search, those websites that use "got it" and equivalent two word synnonyms
#split now in the first space to use optimized comparing set functions as set's would remove spaces
#Problem for decline, those buttons that have "don't allow" and similar using two words, often with one being 
#associated with accept. example "only allow essential and allow all cookies [instagram]"

#Due to these problems may be better to make pair of words then compare pairs
stringlist = []
accept_scrn_list = []
decline_scrn_list = []
accept_notice_list = []
decline_notice_list = []

#due to some sites forcing the site in norwegian regardless of what language is set if the user is from norway
#added norwegian synonyms as well
accepts = np.array(["accept","yes", "allow", "confirm","approve","enable", "consent", "ok", "agree", "godta",
                    "godkjenn", "bekreft", "godtaalle", "tillat", "Got it", "sure", "got it", "gotit"])
declines = np.array(["decline", "deny","refuse","reject", "disagree", "refuse,", "no", "disallow", "avslå",
                    "avvis", "avvisalle"])

#Add better language filter so that "accept," and equivalent is cathced 
for i in overview['notice_screen']:
    im = Image.open(i)
    widthI, heightI = im.size
    widthI = int(widthI)
    heightI = int(heightI)
    if(widthI > 30000 or heightI > 30000):
        im = im.resize((int(widthI/2), int(heightI/2)))
        im = pytesseract.image_to_string(im, config='--psm 4')
        im = im.lower().split()
    else:
        im = pytesseract.image_to_string(im, config='--psm 6')
        im = im.lower().split()
    reg = re.findall("\w+", str(im))
    #print(reg)
    
    #s = re.sub(r'\W+', '', str(s))
    setcheck = np.array([reg])
    #print(s)
   # for i in range(len(setcheck)):

   #     setcheck[i] = str(setcheck[i]).replace(',','')
    
    #to check wether there are some accept synonyms
    found_accept = np.intersect1d(setcheck, accepts, assume_unique=False)
    accept_scrn_list.append(found_accept)
    #To check wether there are some decline synonyms
    found_decline = np.intersect1d(setcheck, declines, assume_unique=False)
    decline_scrn_list.append(found_decline)
    stringlist.append(reg)
#For the text we gathered from the notice we also check that it does not contain any accept or refuse options
for i in overview['notice_text']:
    i=i.lower().split()
    notice_text_set = np.array([i])
    notice_text_accept = np.intersect1d(notice_text_set, accepts, assume_unique=False)
    notice_text_decline = np.intersect1d(notice_text_set, declines, assume_unique=False)
    accept_notice_list.append(notice_text_accept)
    decline_notice_list.append(notice_text_decline)
    
overview['notice_scr_accept'] = accept_scrn_list
overview['notice_scr_decline'] = decline_scrn_list
overview['notice_scr_text'] = stringlist

overview['notice_text_accept'] = accept_notice_list
overview['notice_text_decline'] = decline_notice_list
#For checking wether what type of buttons they contain, more optimized to use sets such that we
#can use numpy common elements function between set of accept synonyms and sets of the notce, same for decline
#and learn more.

#print(overview.head(9))
```
